```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Life Expectancy Data.csv')
df.head()
```

```python
# Checking for Null Values
df.isnull().sum()
```

```python
# Replacing the Null Values with mean values of the data
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)
df['Life expectancy ']=imputer.fit_transform(df[['Life expectancy ']])
df['Adult Mortality']=imputer.fit_transform(df[['Adult Mortality']])
df['Hepatitis B']=imputer.fit_transform(df[['Hepatitis B']])
df['Polio']=imputer.fit_transform(df[['Polio']])
df['Total expenditure']=imputer.fit_transform(df[['Total expenditure']])
df['Diphtheria ']=imputer.fit_transform(df[['Diphtheria ']])
df['GDP']=imputer.fit_transform(df[['GDP']])
df['Population']=imputer.fit_transform(df[['Population']])
df['Schooling']=imputer.fit_transform(df[['Schooling']])
```

```python
# Checking again for Null Values
df.isnull().sum()
```

```python
# iterating the columns
for col in df.columns:
    print(col)
```

```python
# drop unecessary columns 
df = df.drop(columns=['Alcohol', 'Measles ',' BMI ','under-five deaths ',' HIV/AIDS',' thinness  1-19 years',' thinness 5-9 years','Income composition of resources'])
```

```python
# Example
continent = []
df= df.replace("Bolivia (Plurinational State of)", "Bolivia")
df = df.replace("Iran (Islamic Republic of)","Iran")
df = df.replace("Micronesia (Federated States of)","Micronesia")
df = df.replace("Republic of Korea","South Korea")
df = df.replace("The former Yugoslav republic of Macedonia", "Macedonia")
df = df.replace("Timor-Leste",'East Timor')
df = df.replace("Venezuela (Bolivarian Republic of)","Venezuela")
for i in range(0,len(df)):
  country_name = df[i]
  if i>=2601 and i<=2616:
    continent.append("Asia")
  else:
    continent_name = country_to_continent(country_name)
    continent.append(continent_name)
    print(continent_name)
    #print(i)
```
